```python
#DASHBOARD CODE
#-----------------------------------------------------------------------------------------------------------------------------------------------------------


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


plt.style.use('dark_background')
sns.set_style("darkgrid")
plt.rcParams["font.family"] = "DejaVu Sans"


fig = plt.figure(figsize=(20, 18), facecolor='#1a1a1a')  
fig.patch.set_facecolor('#1a1a1a')

plt.suptitle("▣ QUICK COMMERCE DASHBOARD", fontsize=36, color='white', y=0.99, fontweight='bold')
plt.figtext(0.5, 0.95, "▶ Business Overview - Revenue - Delivery - Experience", 
           ha="center", fontsize=18, color='lightgray', style='italic')


Total_Order = df["Order_ID"].count()
Total_Revenue = df["Order_Value"].sum()
Avg_Delivery_Time = df["Delivery_Time_Min"].mean()
Avg_Customer_Rating = df["Customer_Rating"].mean()

plt.subplot2grid((11,4),(0,0))
plt.axis('off')
plt.text(0.5, 0.7, f"{Total_Order:,}", fontsize=32, ha="center", color='#00d4ff', fontweight='bold')  
plt.text(0.5, 0.4, "■ Total Orders", fontsize=18, ha="center", fontstyle="oblique", fontweight="bold", color='white')

plt.subplot2grid((11,4),(0,1))
plt.axis('off')
plt.text(0.5, 0.7, f"${Total_Revenue:,.0f}", fontsize=32, ha="center", color='#00ff88', fontweight='bold')  
plt.text(0.5, 0.4, "$ Total Revenue", fontsize=18, ha="center", fontstyle="oblique", fontweight="bold", color='white')

plt.subplot2grid((11,4),(0,2))
plt.axis('off')
plt.text(0.5, 0.7, f"{Avg_Delivery_Time:.1f} min", fontsize=32, ha="center", color='#ff8c42', fontweight='bold')  
plt.text(0.5, 0.4, "⟲ Avg Delivery Time", fontsize=18, ha="center", fontstyle="oblique", fontweight="bold", color='white')

plt.subplot2grid((11,4),(0,3))
plt.axis('off')
plt.text(0.5, 0.7, f"{Avg_Customer_Rating:.1f} ★", fontsize=32, ha="center", color='#ffd700', fontweight='bold')  
plt.text(0.5, 0.4, "★ Avg Rating", fontsize=18, ha="center", fontstyle="oblique", fontweight="bold", color='white')


company_metrics = df.groupby("Company").agg(
    orders=("Order_ID","count"),
    Revenue=("Order_Value","sum"),
    Avg_Delivery_Time=("Delivery_Time_Min","mean"),
    Avg_Rating=("Customer_Rating","mean")
).reset_index()

ax1 = plt.subplot2grid((9,2),(1,0), rowspan=3)
professional_colors = ['#667eea', '#764ba2', '#f093fb', '#f5576c']  
bars1 = ax1.bar(company_metrics["Company"], company_metrics["orders"], color=professional_colors, 
                edgecolor='white', linewidth=1.5, alpha=0.9)
ax1.set_title("▲ Orders By Company", fontsize=24, color='white', fontweight='bold', pad=20)
ax1.tick_params(axis='x', rotation=45, labelsize=16, colors='white')
ax1.tick_params(axis='y', labelsize=16, colors='white')
ax1.set_xlabel("Company", color='white', fontweight='bold', fontsize=18)
ax1.set_ylabel("Orders", color='white', fontweight='bold', fontsize=18)
ax1.grid(True, alpha=0.2, axis='y')

for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{int(height)}', ha='center', va='bottom', color='black', fontweight='bold', fontsize=12)
             
ax2 = plt.subplot2grid((9,2),(1,1), rowspan=3)
ax2.plot(company_metrics['Company'], company_metrics['Revenue'], marker='o', 
         linewidth=4, markersize=12, color='#ff6b6b', markerfacecolor='#ff4757', 
         markeredgecolor='white', markeredgewidth=2)
ax2.fill_between(company_metrics['Company'], company_metrics['Revenue'], 
                 alpha=0.3, color='#ff6b6b', interpolate=True)
ax2.set_title("$ Revenue By Company", fontsize=24, color='white', fontweight='bold', pad=20)
ax2.tick_params(axis='x', rotation=45, labelsize=16, colors='white')
ax2.tick_params(axis='y', labelsize=16, colors='white')
ax2.set_xlabel("Company", color='white', fontweight='bold', fontsize=18)
ax2.set_ylabel("Revenue", color='white', fontweight='bold', fontsize=18)
ax2.grid(True, alpha=0.3, linestyle='--')

for i, (company, revenue) in enumerate(zip(company_metrics['Company'], company_metrics['Revenue'])):
    ax2.annotate(f'${revenue:,.0f}', (i, revenue), textcoords="offset points", 
                xytext=(0,15), ha='center', color='black', fontweight='bold', fontsize=10)

ax3 = plt.subplot2grid((9,2),(6,0), rowspan=3)
bright_colors = ['#4ecdc4', '#45b7d1', '#96ceb4', '#feca57']  
bars3 = ax3.barh(company_metrics['Company'], company_metrics['Avg_Delivery_Time'], 
                color=bright_colors, edgecolor='white', linewidth=1.5, alpha=0.9)
ax3.set_title("◐ Avg Delivery Time By Company", fontsize=24, color='white', fontweight='bold', pad=20)
ax3.tick_params(axis='y', labelsize=16, colors='white')
ax3.tick_params(axis='x', labelsize=16, colors='white')
ax3.set_ylabel("Company", color='white', fontweight='bold', fontsize=18)
ax3.set_xlabel("Delivery Time (min)", color='white', fontweight='bold', fontsize=18)
ax3.grid(True, alpha=0.2, axis='x')

for i, bar in enumerate(bars3):
    width = bar.get_width()
    ax3.text(width + 0.5, bar.get_y() + bar.get_height()/2, 
             f'{width:.1f}', ha='left', va='center', color='black', fontweight='bold', fontsize=12)


ax4 = plt.subplot2grid((9,2),(6,1), rowspan=3)

num_companies = len(company_metrics['Company'])

scatter_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc', '#99ffcc', '#ffff99', '#cc99ff']

colors_to_use = (scatter_colors * ((num_companies // len(scatter_colors)) + 1))[:num_companies]

scatter = ax4.scatter(range(len(company_metrics['Company'])), company_metrics['Avg_Rating'], 
                     c=colors_to_use, s=200, alpha=0.8, 
                     edgecolors='black', linewidth=3)
ax4.set_title("★ Avg Rating By Company", fontsize=24, color='white', fontweight='bold', pad=20)
ax4.set_xticks(range(len(company_metrics['Company'])))
ax4.set_xticklabels(company_metrics['Company'], rotation=45, ha='right')
ax4.tick_params(axis='x', labelsize=16, colors='white')
ax4.tick_params(axis='y', labelsize=16, colors='white')
ax4.set_xlabel("Company", color='white', fontweight='bold', fontsize=18)
ax4.set_ylabel("Average Rating", color='white', fontweight='bold', fontsize=18)
ax4.grid(True, alpha=0.3, linestyle='--')

for i, (company, rating) in enumerate(zip(company_metrics['Company'], company_metrics['Avg_Rating'])):
    ax4.annotate(f'{rating:.0f}', (i, rating), textcoords="offset points", 
                xytext=(0,15), ha='center', color='black', fontweight='bold', fontsize=12)
```
